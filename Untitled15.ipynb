```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
df = pd.read_csv('dados_vendas.csv')
display(df)
# Converter a coluna de data para o formato correto
df['Data'] = pd.to_datetime(df['Data'])

# Calcular estatsticas bsicas
media_vendas = df['Vendas'].mean()
maior_venda = df['Vendas'].max()
data_maior_venda = df.loc[df['Vendas'].idxmax(), 'Data']
display(df)
# Criar grfico de vendas
plt.figure(figsize=(10,5))
plt.plot(df['Data'], df['Vendas'], marker='o', linestyle='-', color='b', label='Vendas')
plt.axhline(media_vendas, color='r', linestyle='--', label=f'Mdia: {media_vendas:.2f}')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.title('Anlise de Vendas Dirias')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Insight
print(f"A mdia de vendas foi de {media_vendas:.2f} unidades/dia.")
print(f"O dia com maior venda foi {data_maior_venda.strftime('%Y-%m-%d')} com {maior_venda} unidades.")
```
